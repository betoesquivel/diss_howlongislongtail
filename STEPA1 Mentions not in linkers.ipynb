## 1 Doc visualization with pandas

```python
as_item = lambda v: v[0] if isinstance(v, list) else v

tagme_important_info = lambda t: t['title']
# tagme_important_info = lambda t: {'title': t['title'], 'types': t['dbpedia_categories'][:1]}
represent_tagme = lambda t: json.dumps(tagme_important_info(t)) if t.get('spot', None) else None

wikifier_important_info = lambda t: t['title']
# wikifier_important_info = lambda t: {'title': t['title'], 'types': t['types'][:1]}
represent_wikifier = lambda t: json.dumps(wikifier_important_info(t)) if t.get('types', None) else None

stanford_important_info = lambda t: {'count': t['counts']}
represent_stanford = lambda t: json.dumps(stanford_important_info(t)) if t.get('counts', None) else None

tagger_representation = lambda t: filter(lambda r: r is not None, [represent_stanford(t), represent_wikifier(t), represent_tagme(t)])[0]

panda_entity_dict = lambda d: {k: tagger_representation(as_item(v)) for (k,v) in d.items()}
```
